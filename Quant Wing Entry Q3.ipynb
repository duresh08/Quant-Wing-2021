```python
import pandas as pd
import yfinance as yf

Tickers = ['^IXIC','AAPL','MSFT','AMZN','GOOG','FB','TSLA','GOOGL','NVDA','PYPL','ADBE','CMCSA','CSCO','NFLX','INTC',
          'PEP','AVGO','COST','MRNA','TMUS','TXN']
```

```python
data = pd.DataFrame()

for i in Tickers:
    ticker = yf.Ticker('{}'.format(i))
    temp = ticker.history(period = '1y',interval = '1d')
    data = pd.concat([data,temp.iloc[:,3]],axis = 1)
    
data.columns = Tickers
data
```

```python
returns = ((data-data.shift(1))/data.shift(1))*100
returns
```

```python
import scipy
from scipy import stats

Beta = []

for i in range(returns.shape[1]):
    beta,alpha,r_value,p_value,std_error = stats.linregress(returns.iloc[1:,i],returns.iloc[1:,0])
    Beta.append(beta)
    
Beta
```

```python
dictionary = dict(zip(Tickers, Beta))
df = pd.DataFrame(dictionary,index = range(0,22))
df = df.drop(df.index[range(1,df.shape[0])])
df
```

```python
df1 = df.transpose()
df1
```
